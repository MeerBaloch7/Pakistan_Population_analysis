```python
# HOW MUCH RURAL POOPULATION INCRESED FROM 1998-2017
rural_2017=data['ALL SEXES (RURAL)'].sum()
rural_1998=data['POPULATION 1998 (RURAL)'].sum()
change_pop=rural_2017-rural_1998
change_pop
```

```python
print("The rural Population of pakistan have been incresed as compared to 2017 as compared to 1998 :",change_pop)
```

```python
# percentage of increased population
per_of_pp=rural_2017/rural_1998*100
print("Percentage of increased population is: ",per_of_pp)
```
